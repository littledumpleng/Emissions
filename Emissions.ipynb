```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
emissions = pd.read_csv('emissions_estimates2.csv')
print(emissions)
```

```python
emissions.plot(kind='scatter', y='Projected Emissions - Business As Usual', x='Year')
plt.title('Maryland\'s Projected Emissions Over Time - Business as Usual')
plt.xlabel('Year')
plt.ylabel('Emissions')
```

```python
emissions.plot(kind='bar', x='Year', y=['Projected Progress under Current Programs', 'Projected Progress under Enhanced Programs'])
plt.title('Projected Greenhouse Gas Emissions Under Current vs Enhanced Programs')
plt.xlabel('Year')
plt.ylabel('Emissions Estimates')
plt.show()
```

```python
emissions.plot(y='Estimated Emissions', x='Year')
plt.xlabel('Year')
plt.ylabel('Estimated Emissions')
plt.title('Estimated Emissions Between 2006-2014')
plt.show()
```
